```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
url = "https://bigbangtrans.wordpress.com/series-1-episode-1-pilot-episode/"
```

```python
# gets the response from the website
response = requests.get(url)
```

```python
fileLinkEpisodes = ''
soup = BeautifulSoup(response.text, "lxml")
for link in soup.find_all('a'):
    text = link.getText()
    if "Series " in text and " Episode " in text:
        sNum_start = text.find("Series ") + 7
        s = text[sNum_start:sNum_start+2]
        eNum_start = text.find("Episode ") + 8
        e = text[eNum_start:eNum_start+2]
        fileLinkEpisodes += link.get('href') + ' ' + s + 'x' + e + '\n'
print(fileLinkEpisodes)
```

```python
#open text file
link_file = open("./Datasets/Sources/TBBT/fileLink.txt", "w")

#write string to file
link_file.write(fileLinkEpisodes)

#close file
link_file.close()
```

```python
eps = fileLinkEpisodes.split('\n')
epsList = []
for e in eps:
    epsList.append(e.split(' '))
epsList = epsList[:-1]
```

```python
for e in tqdm(epsList):
    script = []
    response = requests.get(e[0])
    soup = BeautifulSoup(response.text, "lxml")
    for div in soup.find_all('div'):
        divClass = div.get('class')
        if ['entrytext'] == divClass:
            for p in soup.find_all('p'):
                script.append(p.getText())
    script = script[:-3]
    episodeName = e[1] + '.txt'
    #open text file
    episode_file = open("./Datasets/Sources/TBBT/Episodes/"+episodeName, "w")
    #write string to file
    filecontent = ''
    for line in script:
        filecontent += line + '\n'
    episode_file.write(filecontent)
    #close file
    episode_file.close()
```
