```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
url_list = ["https://transcripts.foreverdreaming.org/viewforum.php?f=177",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&sid=103afbbe0cbcdc5a83df1dd9e9b032d4&start=25",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=50",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=75",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=100",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=125",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=150",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=175",
            "https://transcripts.foreverdreaming.org/viewforum.php?f=177&start=200"]
```

```python
response_list = []
# gets the response from the website
for url in url_list:
    response_list.append(requests.get(url))
```

```python
fileLinkEpisodes = ''
for response in response_list:
    soup = BeautifulSoup(response.text, "lxml")
    for link in soup.find_all('a'):
        text = link.getText()
        if " - " in text:
            sNum_start = text.find("x") -2
            s = text[sNum_start:sNum_start+2]
            eNum_start = sNum_start + 3
            e = text[eNum_start:eNum_start+2]
            href = link.get('href')
            href = "https://transcripts.foreverdreaming.org/"+href[1:]
            fileLinkEpisodes += href+ ' ' + s + 'x' + e + '\n'
print(fileLinkEpisodes)
```

```python
#open text file
link_file = open("./Datasets/Sources/HIMYM/fileLink.txt", "w")

#write string to file
link_file.write(fileLinkEpisodes)

#close file
link_file.close()
```

```python
eps = fileLinkEpisodes.split('\n')
epsList = []
for e in eps:
    epsList.append(e.split(' '))
epsList = epsList[:-1]
```

```python
for e in tqdm(epsList):
    script = []
    response = requests.get(e[0])
    soup = BeautifulSoup(response.text, "lxml")
    for div in soup.find_all('div'):
        divClass = div.get('class')
        if ['postbody'] == divClass:
            for p in soup.find_all('p'):
                script.append(p.getText())
    script = script[1:-3]
    episodeName = e[1] + '.txt'
    #open text file
    episode_file = open("./Datasets/Sources/HIMYM/Episodes/"+episodeName, "w")
    #write string to file
    filecontent = ''
    for line in script:
        filecontent += line + '\n'
    episode_file.write(filecontent)
    #close file
    episode_file.close()
```
